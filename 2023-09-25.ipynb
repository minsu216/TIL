```python
X = np.linspace(-np.pi, np.pi, 256)
C, S = np.cos(X), np.sin(X)
plt.title("legend를 표시한 플롯")
plt.plot(X, C, ls="--", label="cosine")
plt.plot(X, S, ls=":", label="sine")
plt.legend(loc=2)
plt.show()
```

```python
# 종류별 데이터 건수를 시각화 한다고 한다 . Ex ) bar그래프 사용
```

```python
y = [2, 3, 1]
x = np.arange(len(y)) # 0,1,2
xlabel = ['가', '나', '다']
plt.title("Bar Chart")
plt.bar(x, y)
plt.xticks(x, xlabel) # x축에 해당하는 각각의 차트 bar하나에 대한 이름!!
plt.xlabel('종류')
plt.ylabel('빈도수')
plt.yticks(y)
```

```python
np.random.seed(0)

people = ['몽룡', '춘향', '방자', '향단']
y_pos = np.arange(len(people))
performance = 3 + 10 * np.random.rand(len(people)) # 함수를 통해 4개의 난수가 발생
error = np.random.rand(len(people))

plt.title("Barh Chart")
plt.barh(y_pos, performance, xerr=error, alpha=0.4)
plt.yticks(y_pos, people)
plt.xlabel('x 라벨')
```

```python
x = np.linspace(0.1, 2 * np.pi, 10)
plt.title("Stem Plot")# x데이터가 0.1 ~ 2pi 까지 10개의 구간으로 나눈 값!
plt.stem(x,np.cos(x), '-.') #x데이터에 각각 코사인 데이터 
plt.show()
```

```python
labels = ['개구리', '돼지', '개', '통나무']
sizes = [15, 30, 45, 10]
colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral']
explode = (0, 0.1, 0, 0)
plt.title("Pie Chart")
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=90,counterclock=False,wedgeprops={'edgecolor':\
        'w','width':0.7,'linewidth':5})
#explode : 돌출 여부를 나타냄,
plt.axis('equal')
```
