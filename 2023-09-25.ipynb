```python
X = np.linspace(-np.pi, np.pi, 256)
C, S = np.cos(X), np.sin(X)
plt.title("legend  ")
plt.plot(X, C, ls="--", label="cosine")
plt.plot(X, S, ls=":", label="sine")
plt.legend(loc=2)
plt.show()
```

```python
np.random.seed(0)

people = ['', '', '', '']
y_pos = np.arange(len(people))
performance = 3 + 10 * np.random.rand(len(people)) #   4  
error = np.random.rand(len(people))

plt.title("Barh Chart")
plt.barh(y_pos, performance, xerr=error, alpha=0.4)
plt.yticks(y_pos, people)
plt.xlabel('x ')
```

```python
x = np.linspace(0.1, 2 * np.pi, 10)
plt.title("Stem Plot")# x 0.1 ~ 2pi  10   !
plt.stem(x,np.cos(x), '-.') #x    
plt.show()
```

```python
labels = ['', '', '', '']
sizes = [15, 30, 45, 10]
colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral']
explode = (0, 0.1, 0, 0)
plt.title("Pie Chart")
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=90,counterclock=False,wedgeprops={'edgecolor':\
        'w','width':0.7,'linewidth':5})
#explode :   ,
plt.axis('equal')
```
