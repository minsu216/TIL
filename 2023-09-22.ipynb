```python
#iris Data에 대해서 petal_length값 기준으로 내림차순 정렬을 하고싶아아 
```

```python
iris.groupby(iris.species).get_group('setosa').sort_values(by='petal_length',ascending=False)
def myfunc2(df):
    return df.sort_values(by='petal_length',ascending=False)

iris.groupby(iris.species).apply(myfunc2)
```

```python
def q3cut(s):
    return pd.qcut(s,3,labels=['소','중','대']).astype(str)

#transform은 자체적으로 데이터가 바뀐다는 특징을 가지고 있다!
iris['petal_length_class']=iris.groupby(iris.species).petal_length.transform(q3cut)#iris를 종별로 그룹화하고 각 열에 대해 petal_length열을 가지고 소,중,대로 변환해서 str로 넣어라 !
```

```python
iris[['petal_length','petal_length_class']].tail(10)
```
